```python
import os
import shutil
from sklearn.model_selection import train_test_split
```

```python
split_dir = "/home/anna/skola/KSY/dataset_pismena/split"
train_dir = "/home/anna/skola/KSY/dataset_pismena/split_train_128"
test_dir = "/home/anna/skola/KSY/dataset_pismena/split_test_128"
```

## Create train and test dataset from my dataset

```python
if not os.path.exists(train_dir):
    os.makedirs(train_dir)

if not os.path.exists(test_dir):
    os.makedirs(test_dir)

# Get the list of all files in the split directory
all_files = os.listdir(split_dir)
print(all_files)
```

```python
# Group the files by letter
letter_to_files = {}
for file in all_files:
    letter = file.split('_')[0]
    if letter not in letter_to_files:
        letter_to_files[letter] = []
    letter_to_files[letter].append(file)

# Split and move the files
for letter, files in letter_to_files.items():
    train_files, test_files = train_test_split(files, test_size=0.2, random_state=42)
    
    for file in train_files:
        source_path = os.path.join(split_dir, file)
        target_path = os.path.join(train_dir, file)
        shutil.copy(source_path, target_path)
    
    for file in test_files:
        source_path = os.path.join(split_dir, file)
        target_path = os.path.join(test_dir, file)
        shutil.copy(source_path, target_path)
```

# Resize my dataset from 128x128 to 28x28

```python
import numpy as np
import cv2
```

```python
train_input_dir = "/home/anna/skola/KSY/dataset_pismena/split_train_128"
train_output_dir = "/home/anna/skola/KSY/dataset_pismena/split_train_28"
os.makedirs(train_output_dir, exist_ok=True)

test_input_dir = "/home/anna/skola/KSY/dataset_pismena/split_test_128"
test_output_dir = "/home/anna/skola/KSY/dataset_pismena/split_test_28"
os.makedirs(test_output_dir, exist_ok=True)
```

```python
def load_and_resize_images(input_dir, output_dir, size=(28, 28)):
    # Loop through the images in the input directory
    for filename in os.listdir(input_dir):
        if filename.endswith(".png"):
            # Load the image
            input_image = cv2.imread(os.path.join(input_dir, filename))

            # Resize the image to 28x28
            output_image = cv2.resize(input_image, size)

            # Save the resized image to the output directory
            output_filename = os.path.splitext(filename)[0] + "_28.png"
            output_path = os.path.join(output_dir, output_filename)
            cv2.imwrite(output_path, output_image)
    print(f"Conversion {output_filename} complete")
```

```python
load_and_resize_images(train_input_dir, train_output_dir)
load_and_resize_images(test_input_dir, test_output_dir)
```

## Example from EMNIST

```python
def get_images_randomly(images, labels, count):
    total_images = images.shape[0]
    random_indices = np.random.choice(total_images, count, replace=False)

    random_images = images[random_indices]
    random_labels = labels[random_indices]
    return random_images, random_labels

def show_random_images(images, labels, count):
    random_images, _ = get_images_randomly(images, labels, count)
    for i in range(count):
        image = random_images[i].reshape((28, 28))
        plt.imshow(image, cmap = 'gist_gray')
        plt.show()
```

```python
show_random_images(images_train, labels_train, 30)
```

```python
show_random_images(images_test, labels_test, 30)
```

```python
def invert_image_colors(images):
    inverted_images = np.zeros_like(images)

    for i in range(len(images)):
        image = images[i]
        inverted_image = 255 - image
        inverted_images[i] = inverted_image
        
    return inverted_images
```

```python
emnist_images_train_inverted = invert_image_colors(emnist_images_train)
emnist_images_test_inverted = invert_image_colors(emnist_images_test)
```
